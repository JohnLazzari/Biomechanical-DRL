```python
import numpy as np
import pickle
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, '../')
import config

parser = config.config_parser()
args, unknown = parser.parse_known_args()
```

```python
#Name of the files used for saving train and test kinematics

train_cond_tags = ['c1', 'c2', 'c3', 'c4']
test_cond_tags = ['c1_test', 'c2_test']
```

Save the experimental kinematics in the following format

dict{   

        <'marker_names'> : <['marker_name_1', ..., 'marker_name_n']>,
        
        <'train'> : <dict_train>,
        
        <'test'> : <dict_test>
}

dict_train and dict_test are Python dict objects with following specifications:

dict{ <condition_index(int)>: <kinematics(numpy array with given specs below)>}

condition_index: integer index of the current condition, starts from 0

kinematics: numpy array with following shape: num_markers/targets, num_coordinates =  3, timepoints

num_markers/targets: number of markers/targets recorded for that condition

num_coordinates = 3: x[-->], y[â†‘], z[out of page] coordinates for the target. Values of np.nan will lock that coordinate.

timepoints: total timepoints in that condition

```python
#Create a dictionary object to save later
dict_to_save = {}
```

```python
#First append the markers names to the dict_to_save
#In the monkey cycling experiment only the hand kinematics are recorded
#The marker name must correspond to a body name in the musculoskeletal xml file

marker_names = ['hand']
dict_to_save['marker_names'] = marker_names
```

```python
#Now append the kinematics for the training condition to the dict_to_save
dict_to_save['train'] = {}

for idx_cond, cond in enumerate(train_cond_tags):
    
    x_coord = np.load(f'./x_coord_{cond}.npy').reshape(-1)
    y_coord = np.load(f'./y_coord_{cond}.npy').reshape(-1)
    z_coord = np.ones_like(x_coord)*np.nan #As we do not record/have the z coord of the target, lock that coordinate

    cond_kinematics = np.array([x_coord, y_coord, z_coord])
    
    #As there is just one marker/target, expand the corresponding dimension
    #Otherwise append the kinematics for other markers in that dimension
    cond_kinematics = cond_kinematics[None, ...]
    
    print(cond_kinematics.shape)
    dict_to_save['train'][idx_cond] = cond_kinematics
    
    #Plot the kinematics for sanity check
    plt.plot(cond_kinematics[0, 0])
    plt.plot(cond_kinematics[0, 1])
    plt.plot(cond_kinematics[0, 2])
```

```python
#For sanity check, plot the corresponding kinematics
for cond in range(len(dict_to_save['train'])):
    for markers in range(dict_to_save['train'][cond].shape[0]):
        plt.plot(dict_to_save['train'][cond][markers, 0, :])
        plt.plot(dict_to_save['train'][cond][markers, 1, :])
        plt.plot(dict_to_save['train'][cond][markers, 2, :])
```

```python
#Now append the experimental kinematics for the testing conditions
dict_to_save['test'] = {}

for idx_cond, cond in enumerate(test_cond_tags):
    
    x_coord = np.load(f'./x_coord_{cond}.npy').reshape(-1)
    y_coord = np.load(f'./y_coord_{cond}.npy').reshape(-1)
    z_coord = np.ones_like(x_coord)*np.nan #As we do not have/record the z_coord, lock that coordinate
    
    cond_kinematics = np.array([x_coord, y_coord, z_coord])
    
    #As there is just one target, expand the corresponding dimension
    #Otherwise concatenate in that dimension
    cond_kinematics = cond_kinematics[None, ...]
    
    print(cond_kinematics.shape)
    dict_to_save['test'][idx_cond] = cond_kinematics
    
    #Plot the kinematics for sanity check
    plt.plot(cond_kinematics[0, 0])
    plt.plot(cond_kinematics[0, 1])
    plt.plot(cond_kinematics[0, 2])
```

```python
#For sanity check plot the kinematics for the testing conditions
for cond in range(len(dict_to_save['test'])):
    for marker in range(dict_to_save['test'][cond].shape[0]):
        print(cond)
        plt.plot(dict_to_save['test'][cond][marker, 0, :])
        plt.plot(dict_to_save['test'][cond][marker, 1, :])
        plt.plot(dict_to_save['test'][cond][marker, 2, :])
```

```python
print(dict_to_save['marker_names'])
print(dict_to_save['train'].keys())
print(dict_to_save['test'].keys())
```

```python
#create a binary pickle file 
#Since the path is from usim original dir, append '../'

path_to_save = '../' + args.kinematics_path
f = open(path_to_save + '/kinematics.pkl',"wb")

# write the python object (dict) to pickle file
pickle.dump(dict_to_save, f)

# close file
f.close()
```
